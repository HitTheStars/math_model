## 🌸 标题怎么写呀？

在文字前面加上 `#` 和空格，就可以变成标题啦~ #越多字越小哦，就像这样：

# 一级标题 (最大的标题喵~)
## 二级标题
### 三级标题
#### 四级标题
##### 五级标题
###### 六级标题 (最小的标题喵~)

## ✨ 文字要怎么变漂亮？

有时候我们想让文字变粗或者斜体，只要用 `*` 或者 `_` 把文字包起来就好啦~

*斜体文字* 或者 _斜体文字_ (像小猫的尾巴一样弯弯的~)
**粗体文字** 或者 __粗体文字__ (胖乎乎的好可爱~)
***粗斜体文字*** 或者 ___粗斜体文字___ (又胖又弯的文字喵~)

## 📝 列表要怎么排排坐？

### 无序列表
用 `-`、`*` 或者 `+` 加上空格，就可以做出可爱的列表啦：

- 猫咪喜欢的食物
  - 小鱼干🐟
  - 猫罐头🥫
  - 鸡胸肉🍗
* 猫咪讨厌的东西
  * 洗澡🚿
  * 被踩到尾巴😿

### 有序列表
用 `数字.` 加上空格，就会自动排序哦：

1. 起床伸懒腰😼
2. 舔毛洗脸🧼
3. 找主人要零食🍪
4. 趴在键盘上睡觉💤

## 🔗 链接和图片要怎么加？

### 超链接
想放链接的话，就用 `[链接文字](链接地址)` 这样的格式哦：

[猫咪图片网站](https://cataas.com/) (这里有好多可爱的猫咪图片喵~)

### 图片
放图片和链接很像，只要在前面加个 `!` 就好啦：

![可爱猫咪](https://cataas.com/cat)

## 💻 代码要怎么展示？

### 行内代码
用 ` 把代码包起来，像这样：`print('喵~')`

### 代码块
用三个 ` 加上语言名称，就能做出漂亮的代码块啦：

```python
# 这是Python代码哦~
def cat_sound():
    return '喵~ 喵~'

print(cat_sound())  # 输出：喵~ 喵~
```
