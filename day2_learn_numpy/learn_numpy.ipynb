```python
import numpy as np
```

```python
popu=np.array([19571216,
19673200,
19854526,
20104710,
19463131,
19544098,
19593849,
19636391,
19657321,
19653431,
19626488])
```

```python
print(popu[0],popu[-1])
```

```python
change=np.array([])
for i in range(popu.shape[0]):
    if i<popu.shape[0]-1 :
        change=np.append(change,popu[i+1]-popu[i])
print(change)
```

```python
change=np.array([])
for i in range(popu.shape[0]):
    if i<popu.shape[0]-1 :
        change=np.append(change,popu[i+1]-popu[i])
print(change)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 6)) 
plt.title("NYC population change from 1950 to 2020")
plt.xlabel("Year")
plt.ylabel("Population change")
sns.lineplot(y=change,x=np.arange(1, len(change) + 1))
plt.show()
```
